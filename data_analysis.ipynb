```python
#import libs
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#collect dataframes
df_k2 = pd.read_csv('results/k-results/k_2.csv')
df_k4 = pd.read_csv('results/k-results/k_4.csv')
df_k8 = pd.read_csv('results/k-results/k_8.csv')
df_k16 = pd.read_csv('results/k-results/k_16.csv')
```

```python
#TEST R_values
#gather R_values
k2_r_avg_values = df_k2['R_AVG'].to_list()
k2_r_avg_values = np.mean(k2_r_avg_values)

k4_r_avg_values = df_k4['R_AVG'].to_list()
k4_r_avg_values = np.mean(k4_r_avg_values)

k8_r_avg_values = df_k8['R_AVG'].to_list()
k8_r_avg_values = np.mean(k8_r_avg_values)

k16_r_avg_values = df_k16['R_AVG'].to_list()
k16_r_avg_values = np.mean(k16_r_avg_values)
```

```python
#plot graph
k_values = ['k2', 'k4', 'k8', 'k16']
r_avg_means = [k2_r_avg_values, k4_r_avg_values, k8_r_avg_values, k16_r_avg_values]

plt.figure(figsize=(8, 6))
plt.bar(k_values, r_avg_means, color='skyblue')
plt.xlabel('k')
plt.ylabel('Média de R_AVG')
plt.title('Média de R_AVG por k')
plt.show()
```

```python
#TEST number of updates/inserts/deletes
#gather_values
vertex_num = df_k2["VERTEX_NUM"]
edges_num = df_k2["EDGES_NUM"]

updt_list = df_k2['UPDATES']/edges_num
isrt_list = df_k2["INSERTS"]/vertex_num
delt_list = df_k2["DELETES"]/vertex_num
```

```python
plt.figure(figsize=(8, 6))
plt.plot(vertex_num, isrt_list, marker='o', linestyle='-', color='blue')
plt.xlabel('VERTEX_NUM')
plt.ylabel('INSERTS / VERTEX_NUM')
plt.title('INSERTS/vertex_num vs VERTEX_NUM')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(vertex_num, delt_list, marker='o', linestyle='-', color='blue')
plt.xlabel('VERTEX_NUM')
plt.ylabel('DELETES / VERTEX_NUM')
plt.title('DELETES/vertex_num vs VERTEX_NUM')
plt.grid(True)
plt.show()

edges_num_norm = (edges_num - min(edges_num)) / (max(edges_num) - min(edges_num))

# Ordenar os valores pelo eixo X
sorted_indices = np.argsort(edges_num_norm)
edges_num_sorted = edges_num_norm.iloc[sorted_indices]
updt_list_sorted = updt_list.iloc[sorted_indices]

plt.figure(figsize=(10, 6))
plt.plot(edges_num_sorted, updt_list_sorted, marker='o', linestyle='-', color='blue')

# Configurações do gráfico
plt.xlabel('Normalized EDGES_NUM')
plt.ylabel('UPDATES / EDGES_NUM')
plt.title('UPDATES vs EDGES_NUM (X Normalized)')
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
